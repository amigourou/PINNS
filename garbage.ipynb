```python
import torch

# Example data
N, p = 10, 6  # Tensor shape
data = torch.randn(N, p, requires_grad=True)  # Enable gradients for autograd

# Compute gradient along dimension 0 for each column in data
grads = []
for col in range(p):
    grad_col = torch.autograd.grad(outputs=data[:, col].sum(), inputs=data, retain_graph=True, create_graph=True)[0][:, col]
    grads.append(grad_col)

# Stack gradients along the second dimension to match original shape
gradient_tensor = torch.stack(grads, dim=1)

print("Original data shape:", data.shape)
print("Gradient tensor shape:", gradient_tensor.shape)
```
